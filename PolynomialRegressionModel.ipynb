```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Genrate same synthetic dp or reproducibilty of results

np.random.seed(1)
```

```python
# Genrate synthetic x 

X = 2*np.random.rand(100 ,1)
```

```python
X
```

```python
# Genrate random synthetic y using X

y = 4 + 3*X+1.5*X**2 + np.random.randn(100, 1)
```

```python
y

```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y, test_size=0.2 , random_state=1)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
# perform polynomial regeression 

from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_feature = PolynomialFeatures(degree=2 , include_bias=False)
X_poly_train = poly_feature.fit_transform(X_train)
```

```python
X_poly_train
```

```python
from sklearn.linear_model import LinearRegression
Poly_Reg = LinearRegression()
```

```python
Poly_Reg
```

```python
Poly_Reg.fit(X_poly_train , y_train)
```

```python
Poly_Reg.coef_
```

```python
Poly_Reg.intercept_
```

```python
#make prediction on train data itself for the evaluation metrics
y_poly_predict = Poly_Reg.predict(X_poly_train)
```

```python
y_poly_predict
```

```python
from sklearn.metrics import  mean_squared_error , r2_score
print(mean_squared_error(y_train , y_poly_predict))
print(r2_score(y_train , y_poly_predict))
```

```python
# Now make prediction on test data 

X_poly_test = poly_feature.transform(X_test)
```

```python
y_test_pred = Poly_Reg.predict(X_poly_test)
```

```python
# Evaluation metrics

print(mean_squared_error(y_test , y_test_pred))
print(r2_score(y_test , y_test_pred))
```

```python
# graphical representaion 
degree = 2
plt.scatter(X_test , y_test , label= 'Testing data ')
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_poly = poly_feature.transform(X_range)
plt.plot(X_range, Poly_Reg.predict(X_range_poly), color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
